## Grayscale


```python
type(img)
```

```python
import numpy as np

def grayscale(img):

    r, g, b = img[:, :, 0], img[:, :, 1], img[:, :, 2]

    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
img2 = grayscale(img)
img2
```

```python
from PIL import Image as im
from matplotlib import pyplot as plt

plt.imshow(im.fromarray(img2), interpolation='nearest')
plt.show()
```

## Binary

```python
img2.shape
```

```python
def binary(img):
    result = img

    for i in range(len(img2)):
        for j in range(len(img2[i, :])):
            if img[i, j] > 127:
                result[i, j] = 255
            else:
                result[i,j ] = 0

    return result
```

```python
img3 = binary(img2)
img3
```

```python
plt.imshow(im.fromarray(img3), interpolation='nearest')
plt.show()
```
